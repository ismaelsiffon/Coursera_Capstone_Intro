```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
source= requests.get('https://en.wikipedia.org/wiki/List_of_districts_in_the_London_Borough_of_Redbridge').text
soup= BeautifulSoup(source,'lxml')
liste= soup.find('ul')
#print(liste.prettify())
```

```python
redbridge_neigh=[]
for li in liste.find_all('li'):
    try:
        neigh= li.a.text
    except:
        neigh= li.text
    redbridge_neigh.append(neigh)
redbridge_neigh
```

```python
redbridge_neigh_df= pd.DataFrame(columns=['Neighborhood','Latitude','Longitude'])
redbridge_neigh_df
```

```python
import googlemaps
API_key= 'AIzaSyAaZ45aH6Iuhjkx8reDnDDZ1qQ5Cs61h2g'
gmaps= googlemaps.Client(key= API_key)
```

```python
num_of_neigh= len(redbridge_neigh)
for i in range(0, num_of_neigh):
    neigh= redbridge_neigh[i]
    redbridge_neigh_df.loc[i, 'Neighborhood']= neigh
    results= gmaps.geocode(neigh+', redbridge, london')
    try:
        lat= results[0]['geometry']['location']['lat']
        lon= results[0]['geometry']['location']['lng']
    except:
        lat= 'None'
        lon= 'None'
    redbridge_neigh_df.loc[i, 'Latitude']= lat
    redbridge_neigh_df.loc[i, 'Longitude']= lon
    
        
    
redbridge_neigh_df
```
