```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
source= requests.get('https://en.wikipedia.org/wiki/List_of_districts_in_the_London_Borough_of_Harrow').text
soup= BeautifulSoup(source,'lxml')
table= soup.find('table',class_='wikitable')
#print(table.prettify())
```

```python
harrow_neigh_list= []

for row in table.find_all('tr')[1:]: #loop through rows
        try: #this will exclude the rows that are headers
            col1= row.find_all('td')[0] #we want info from the first column of each row
            neigh= col1.find('a').text # find the text
            harrow_neigh_list.append(neigh)
        except:
            pass
print(len(harrow_neigh_list))
print(harrow_neigh_list)
```

```python
harrow_neigh_df= pd.DataFrame(columns=['Neighborhood','Latitude','Longitude'])
harrow_neigh_df
```
